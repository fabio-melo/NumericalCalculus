```python
from pretty import mat_print, mathjax
from pprint import pprint
from numpy import array, zeros, diag, diagflat, dot
import numpy as np
from IPython.display import display, HTML

def jacobi(A,b,N=25,x=None, tol=1e-255):
    mathtx = mathjax
    
    """Solves the equation Ax=b via the Jacobi iterative method."""
    # Create an initial guess if needed                                                                                                                                                            
    if x is None:
        x = zeros(len(A[0]))
    
    i = 0
    mathtx += mat_print(i,x)
    i += 1

    # Create a vector of the diagonal elements of A                                                                                                                                                
    # and subtract them from A                                                                                                                                                                     
    D = diag(A)
    R = A - diagflat(D)
    x0 = x

    while i < N:
        x = (b - dot(R,x)) / D
        mathtx += mat_print(i,x)
        if np.allclose(x0,x, atol=tol):
            break
        else:
            x0 = x
        i += 1
    
    return mathtx

A = array([[2.0,1.0],[5.0,7.0]])
b = array([11.0,13.0])
guess = array([0,0,0])

M = array([[252,61,113],[2,178,21],[6,1,21]])
B = array([12,8,1])

sol = jacobi(M,B,N=20,x=guess)
display(HTML(sol))
```

```python
from pretty import mat_print, mathjax
from pprint import pprint
from numpy import array, zeros, diag, diagflat, dot



def seidel(a,b,x,t=50, tol=1e-8):
    mathtx = mathjax
    x0 = x.copy()
    n=len(x)
    
    m = 0
    mathtx += mat_print(m,x)
    m += 1
    
    
    while m < t:
       
        for i in range(n):
            s=0
            for j in range(n):
                if i!=j:
                    s += a[i,j] *x[j]
            x[i] = (b[i]-s)/a[i,i]
          
        if all([(False if abs(x0[k] - x[k]) > tol else True) for k in range(len(x))]):                 
            break
        mathtx += mat_print(m,x)
        
        x0 = x.copy()
        m += 1
    return mathtx



guess = array([0.0,0.0,0.0],dtype=np.float32)

A = array([[252,61,113],[2,178,21],[6,1,21]],dtype=np.float32)
B = array([12,8,1], dtype=np.float32)

sol = seidel(A,B,guess, t=200)
display(HTML(sol))
```

```python
from pretty import mat_print, mathjax
from pprint import pprint
from numpy import array, zeros, diag, diagflat, dot
import numpy as np
from IPython.display import display, HTML


def jacobi(a,b,x,t=50, tol=1e-8):
    mathtx = mathjax
    x0 = x.copy()
    n=len(x)
    
    m = 0
    mathtx += mat_print(m,x)
    m += 1
    
    
    while m < t:
        q = np.zeros(x.shape)
        for i in range(n):    
            s=0
            for j in range(n):
                if i!=j:
                    s += a[i,j] *x[j]
            q[i] = ((b[i]-s)/a[i,i])
        x = np.array(q)
        if all([(False if abs(x0[k] - x[k]) > tol else True) for k in range(len(x))]):                 
            break
        mathtx += mat_print(m,q)
        
        x0 = np.array(q)
        m += 1
    return mathtx



guess = array([0.0,0.0,0.0],dtype=np.float32)

A = array([[252,61,113],[2,178,21],[6,1,21]],dtype=np.float32)
B = array([12,8,1], dtype=np.float32)

sol = jacobi(A,B,guess, t=20)
display(HTML(sol))
```
